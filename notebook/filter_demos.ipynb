```python
# Base library imports
import os
import re

# Module imports
import numpy as np
from skimage import img_as_int
from skimage.filters import gaussian
from skimage.io import imshow

# Quick pathfix required before loading project resources
try:
    notebook_location
except NameError:
    notebook_location = os.getcwd()
    os.chdir(re.sub(r'(adversarial-filters)[\\/].*', r'\1', notebook_location))

# Project imports
from filter import ColorBlindnessFilter
from filter import DaltonizationFilter
from filter import DaltonizationLabEnchanceFilter
from filter import FourierEllipsoidFilter
from filter import FourierGaussianFilter
from filter import FourierUniformFilter
from filter import GaussianFilter
from filter import GaussianNoiseFilter
from filter import HistogramEqualizationFilter
from filter import SaltPepperNoiseFilter
from filter import SpeckleNoiseFilter
from filter import WaveletDenoiseFilter
```

```python
images = np.load('data/vgg16_100_correct.npy', allow_pickle=True)
image_index = 79
```

### Original Image

```python
imshow(images[image_index][0]/255.)
```

### Color Blindness Filter

```python
filter = ColorBlindnessFilter()
imshow(filter.transform_image(['prot'], images[image_index][0])/255.)
```

### Daltonization Filter

```python
filter = DaltonizationFilter()
imshow(filter.transform_image(['prot'], images[image_index][0])/255.)
```

### Daltonization and LAB Enhancement

```python
filter = DaltonizationLabEnchanceFilter()
imshow(filter.transform_image(['prot'], images[image_index][0])/255.)
```

### Fourier Ellipsoid

```python
filter = FourierEllipsoidFilter()
imshow(filter.transform_image([5.22821402060977], images[image_index][0])/255.)
```

### Fourier Gaussian

```python
filter = FourierGaussianFilter()
imshow(filter.transform_image([1.], images[image_index][0])/255.)
```

### Fourier Uniform

```python
filter = FourierUniformFilter()
imshow(filter.transform_image([2.], images[image_index][0])/255.)
```

### Gaussian

```python
filter = GaussianFilter()
imshow(filter.transform_image([2.2458103940439136], images[image_index][0])/255.)
```

### Gaussian Noise

```python
filter = GaussianNoiseFilter(454)
imshow(filter.transform_image([-0.05, 0.018884564330270543], images[image_index][0])/255.)
```

### Histogram Equalization

```python
filter = HistogramEqualizationFilter()
imshow(filter.transform_image([256, True], images[image_index][0])/255.)
```

### Salt and Pepper Noise

```python
filter = SaltPepperNoiseFilter(454)
imshow(filter.transform_image([0.05, 0.5], images[image_index][0])/255.)
```

### Speckle Noise

```python
filter = SpeckleNoiseFilter(454)
imshow(filter.transform_image([-0.04986624751630079, 0.0908713646800712], images[image_index][0])/255.)
```

### Wavelet Denoise

```python
filter = WaveletDenoiseFilter()
imshow(filter.transform_image([0.11250517415327012], images[image_index][0])/255.)
```
